# Q2 Create a new column with name 'description' combining `'overview' and 'tagline'` columns in the given dataset

```python
cb_df.isnull().sum()
```

```python
# Lamdba function is created to concatenate the values to check the null values in each column
# ensuring only column with null in both are null - 954 records with null values in both columns
cb_df['description']=''
cb_df['description']=cb_df.apply(lambda x: cb_df['overview'] if cb_df['tagline'].isnull else ( cb_df['overview'] if cb_df['overview'].isnull else cb_df['overview']+cb_df['tagline']) )
```

# Popularity Based Recommendation System

### About Dataset

Anonymous Ratings on jokes.

1. Ratings are real values ranging from -10.00 to +10.00 (the value "99" corresponds to "null" = "not rated").

2. One row per user

3. The first column gives the number of jokes rated by that user. The next 100 columns give the ratings for jokes 01 - 100.

# Q10 Read the dataset(jokes.csv)

Take care about the header in read_csv() as there are no column names given in the dataset. 

```python
j_df = pd.read_csv('jokes.csv',header=None, skiprows=1)
j_df.head()
```

# Q11 Consider `ratings` named dataframe with only first 200 rows and all columns from 1(first column is 0) of dataset

```python
ratings=j_df.iloc[:200,1:]
ratings
```

```python
ratings.shape
```

# Q12 Change the column indices from 0 to 99

```python
ratings.columns
```

```python
colnames=[]
colnames=list(range(100))
ratings.set_axis(colnames,axis=1,inplace=True)
```

```python
ratings.head(1)
```

# Q13 In the dataset, the null ratings are given as 99.00, so replace all 99.00s with 0
Hint: You can use `ratings.replace(<the given value>, <new value you wanted to change with>)`

```python
ratings.replace(99,0,inplace=True)
ratings.head(1)
# Note: For rating i have considered zero as an observation ( not a NaN, so i have not given any treatment)
```

# Q14 Normalize the ratings and save them in `ratings_diff` variable

```python
ratings_diff = (ratings - ratings.mean())
```

```python
ratings_diff
```
