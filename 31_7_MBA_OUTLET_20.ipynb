REMOVING THE TRANSACTION RECORDS WITH DATA MORE THAN 2

```python
transaction_data=item_df[item_df["itemname"].map(len)>2]
```

```python
transactions=np.asarray(transaction_data["itemname"])
```

```python
patterns = pyfpgrowth.find_frequent_patterns(transactions, 10)
```

```python
rules = pyfpgrowth.generate_association_rules(patterns,0.8)
```

```python
rules_df=pd.DataFrame.from_dict(rules,orient='index')
```

```python
rules_df=rules_df.reset_index()
```

```python
rules_df=rules_df.rename(columns={rules_df.columns[0]: 'Antecedent'})
```

```python
rules_df=rules_df.rename(columns={rules_df.columns[1]: 'Consequent'})
```

```python
rules_df=rules_df.rename(columns={rules_df.columns[2]: 'Confidence'})
```

FINDING OUT THE LIFT: CALCULATING THE SUPPORT FOR RHS

```python
def support_count(rhs):
    count=0
    rhs = set(rhs)
    for j in transaction_data['itemname']:
        j = set(j)
        if(rhs.issubset(j)):
            count=count+1
    return count
```

```python
item_count=[]
for i in rules_df["Consequent"]:
    a = support_count(i)
    item_count.append(a/len(transaction_data))
```

```python
item_count
```

```python
rules_df.shape
```

```python
rules_df['RHS_count']=item_count
```

```python
rules_df=rules_df[['Antecedent','Consequent','Confidence','RHS_count']]
```

```python
rules_df
```

```python
rules_df['lift']= rules_df['Confidence']/rules_df['RHS_count']
```

```python
rules_df
```

                                 CALCULATING CONVICTION

```python
rules_df['Conviction']=(1-rules_df['RHS_count'])/(1-rules_df['Confidence'])
```

```python
rules_df.to_csv("rules.csv")
```

```python
rules_df['LHS']= rules_df['Antecedent'].astype(str)
```

```python
rules_df['LHS']=rules_df['LHS'].map(lambda x: x.lstrip('(').rstrip(')'))
```

```python
rules_df['RHS']= rules_df['Consequent'].astype(str)
```

```python
rules_df['RHS']=rules_df['RHS'].map(lambda x: x.lstrip('[').rstrip(']'))
```

```python
rules_df['items_combined_lhs_rhs']=rules_df['LHS'] + ',' +  rules_df['RHS']
```
